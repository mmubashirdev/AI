```python
import random;

def guess_number():
  secret = random.randint(1,10)
  for attempt in range(5):
    guess = int(input("Guess the number from 1 - 10: "))
    if guess == secret:
      print("You guessed the correct number")
      return 
    elif abs(guess - secret) == 1:
      print("You are very near")
    elif guess < secret:
      print("Too Low")
    else:
      print("Too High")
    
  print("You lose the number was ",secret)
  
guess_number()
```

```python
def is_Palindrome(word):
  return word == word[::-1]

words = ["radar","python","madam","world"]

for w in words:
  print(f"{w} is a palindrome: ",is_Palindrome(w))
```

```python
def check_Prime(num):
  if num<2:
    return False
  for i in range(2,num):
    if num % i == 0:
      return False
  return True

for num in range(1,20):
  print(num," is prime: ",check_Prime(num))
   
```
